```python
base_dir = "../../data"
image_dir = f"{base_dir}/preprocessed_images"
csv_file = f'{base_dir}/ODIR-5K/data.csv'

def get_unique_values(file, col):
    unique_diags = set()
    for ix, row in pd.read_csv(file).iterrows():
        left = row[col]
        p = left.split(',')
        for l in p:
            unique_diags.add(l)
    return unique_diags

left = get_unique_values(csv_file, 'Left-Diagnostic Keywords')
right = get_unique_values(csv_file, 'Right-Diagnostic Keywords')
z = right.union(left)

for p in z:
    print (f'{p}')
```
